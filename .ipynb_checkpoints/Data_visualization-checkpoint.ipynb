### Data

```python
fan_reviewed['STARS_DIFF'].count()
```

**I would like to know how many times there is a difference between STAR and RATING:**

**In this graph, we see that one of the movies has a 1 star difference between the displayed rating and the true rating!**

### Ratings from Rotten Tomatoes

Let's start with the ratings from Rotten Tomatoes. There are two sets of ratings here - those published by official critics and user ratings. 

**I'll draw a scatterplot graph showing the relationship between critic and user ratings for Rotten Tomatoes.

```python
plt.figure(figsize = (10,3), dpi = 400)
sns.scatterplot(data = all_sites, x = 'RottenTomatoes', y = 'RottenTomatoes_User')
plt.ylim(0,100)
plt.xlim(0,100)
```

Let's examine the numerical difference between these two ratings. We will calculate this difference using the RottenTomatoes-RottenTomatoes_User formula. The resulting Rotten_Diff values are the critics' rating minus the users' rating. Close to zero values indicate that critics and users agree with each other. Large positive values are when critics rated a movie much higher than users. And large negative values - on the contrary, users rated the movie much higher than the critics.

**Create a new column where I keep the difference between the rating from critics and the user rating for Rotten Tomatoes.**

```python
all_sites['Rotten_Diff'] = all_sites['RottenTomatoes'] - all_sites['RottenTomatoes_User']
all_sites.head()
```

Now let's calculate the average value of the obtained difference over the whole data set. Since the difference can be both positive and negative, let's first take the absolute values of all the values, and only after that let's calculate the average value. In this way, we will get the average of the absolute difference between critic ratings and user ratings.

```python
all_sites['Rotten_Diff'].apply(abs).mean()
all_sites['Abs_Rotten_Diff'] = all_sites['Rotten_Diff'].apply(abs).mean()
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.histplot(data=all_sites,x='Rotten_Diff',kde=True,bins=25)
plt.title("RT Critics Score minus RT User Score");
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.histplot(x = all_sites['Rotten_Diff'].apply(abs),kde=True,bins=25)
plt.title("Abs Difference between RT Critics Score and RT User Score")
```

```python
print("Users Love, but Critics Hate")
all_sites.nsmallest(5, "Rotten_Diff")[['FILM','Rotten_Diff']]
```

```python
print('Critics love, but Users Hate')
all_sites.nlargest(5, 'Rotten_Diff')[['FILM', 'Rotten_Diff']]
```

## Comparing Fandango's ratings to those of other companies

Let's finally find out if Fandango's ratings were artificially high in order to sell more tickets.

**Connect the fandango table with the all_sites table. Not every movie in the fandango table will be found in the all_sites table, because some fandango movies have very few or no reviews. But we only want to compare movies that are in both tables. So I do a "*inner* merge" of the two datasets, matching rows by the FILM.**

```python
df = pd.merge(fandango, all_sites,on='FILM',how='inner')
```

### Normalization of columns to the 0-5 range according to Fandango's STARS and RATINGS columns

Note that Rotten Tomatoes, Metacritic, and IMDB use a different rating measurement scale than Fandango's 0 to 5 star scale. To make comparisons, we need to *normalize* these values to fall within the 0 to 5 star range.

```python
df['RT_Norm'] = df['RottenTomatoes'] /20
df['RTU_Norm'] = df['RottenTomatoes_User'] / 20 
df['Meta_Norm'] = df['Metacritic'] / 20 
df['Meta_U_Norm'] = df['Metacritic_User'] /2
df['IMDB_Norm'] = df['IMDB']/2
```

```python
df['RT_Norm'] = df['RT_Norm'].round(1)
df['RTU_Norm'] = df['RTU_Norm'].round(1)
df['Meta_Norm'] = df['Meta_Norm'].round(1)
df['Meta_U_Norm'] = df['Meta_U_Norm'].round(1)
df['IMDB_Norm'] = df['IMDB_Norm'].round(1)
```

```python
df.head()
```

**Further create a DataFrame named norm_scores, in which I leave only the normalized ratings. I also keep the STARS and RATING columns from the original table fandango.**

```python
norm_scores = df[['STARS', 'RATING','RT_Norm','RTU_Norm','Meta_Norm','Meta_U_Norm','IMDB_Norm']].copy()
norm_scores.head(1)
```
